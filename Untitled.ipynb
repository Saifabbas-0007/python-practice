```python
import matplotlib.pyplot as plt
x =['Mon','Tues','Wed','Thur','Fri','Sat']
y = [10,15,7,12,20,18]
plt.plot(x,y)
plt.title("Bakery Sales This Week !")
plt.xlabel(" Day Of The Week !")
plt.ylabel(" Sales Per Day !")
plt.show()
```

```python
#plt.plot(x,y,color= 'color name',linestyle='linestyle',linewidth= value,marker='marker symbol',label='labelname')
import matplotlib.pyplot as plt
months = [1,2,3,4,5]
sales = [1000,1500,1200,1800,2000]
plt.plot(months,sales,color ='blue',linestyle ='--',linewidth = 2, marker = 'o',label = '2025 sales data')
plt.xlabel('Months')
plt.ylabel('Sales per months')
plt.title('Monthly sales data')
plt.grid(color='black',linestyle= ':',linewidth= 0.5)
plt.xlim(1,6)
plt.ylim(0,2000)
plt.xticks([1,2,3,4,5],['M1','M2','M3','M4','M5'])
plt.show()
```

```python
#plt.bar (x,y,color='color name',width= value,label = 'label name')
import matplotlib.pyplot as plt
Product = ['A','B','C','D']
Sales = [1000,1200,800,1500]
plt.bar(Product,Sales,color= 'orange',label= 'Sales 2025')
plt.xlabel('Sales')
plt.ylabel('Product')
plt.title('Sales Product')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
#plt.pie(values,label= label name,color= color name,autopct='%1.1f%%')
regions= ['North','South','East','West']
revenue= [3000,2000,1500,1000]
plt.pie(revenue, labels= regions, autopct='%1.1f%%', colors= ['gold','lightgreen','skyblue','gray'])
plt.title('Revenue Contribution By Regions')
plt.show()
```

```python
#plt.hist(data,bins=no_of_bins,color= 'colorname',edgecolor='colorname')
import matplotlib.pyplot as plt
scores=[33,45,48,55,89,98,78,77,67,99,47,63,70,80,88,77,66,69,89,]
plt.hist(scores,bins=6,color='gray',edgecolor='black')
plt.xlabel('Scores Range')
plt.ylabel('Number Of Students')
plt.title('Score Distribution Of The Students')
plt.show()
```

```python
#plt.scatter(x,y,color='color name',marker='marker name',label='label name')
import matplotlib.pyplot as plt
Hours_studies= [1,2,3,4,5,6,7,8]
Exam_scores= [50,55,60,65,70,75,80,85,]
plt.scatter(Hours_studies,Exam_scores,color='blue',marker='>',label='Student Data')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Scores')
plt.title('Relationship b/w Study Time & Exam Score')
plt.legend()
plt.grid(True)
plt.show()
```
